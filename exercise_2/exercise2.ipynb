## Plot slope fields and isocline

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
% matplotlib inline

rcParams.update({'font.size': 25})
```

```python
ISO = [-2, -1, 0, 1, 2] #V/s
t = np.linspace(-5, 5, 30)
v = np.linspace(-5, 5, 30)
t_grid , v_grid = np.meshgrid(t, v)
```

```python
ode_rhs = lambda V, t:  np.sin(t) - V /  1.50 # Define  the DGL r i g h t  hand  side
```

```python
lng = np.sqrt(ode_rhs(v_grid, t_grid)**2 + 1) #Lenght of the vector
dt = 1/lng #get the horizontal component
dv = ode_rhs(v_grid, t_grid) / lng #get the vertical component
```

```python
iso_line = {}
for i, iso in enumerate(ISO):
     iso_line[i]= ode_rhs(iso, t)
```

```python
fig, ax = plt.subplots(figsize=(10, 10));
ax.set_ylabel('dV/dt')
ax.set_xlabel('Time (s)')
ax.set_title('')

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
plt.yticks(np.arange(-5, 6, 1));
plt.xticks(np.arange(-5, 6, 1));

plt.quiver(t, v, dt, dv);

# for i, n in enumerate(iso_line):
#      plt.plot(t, iso_line[n], linewidth=2);

iso_plt0 = []
iso_plt00, = plt.plot(t, iso_line[0], linewidth=2, label='-2')
iso_plt01, = plt.plot(t, iso_line[1], linewidth=2, label='-1')
iso_plt02, = plt.plot(t, iso_line[2], linewidth=2, label=' 0')
iso_plt03, = plt.plot(t, iso_line[3], linewidth=2, label=' 1')
iso_plt04, = plt.plot(t, iso_line[4], linewidth=2, label=' 2')
plt.legend(loc=1, handles=[iso_plt00, iso_plt01, iso_plt02, iso_plt03, iso_plt04])
plt.savefig('report/1_1.png', dpi=200)
```

```python
ode_rhs2 = lambda V, t:  1 - V - t 
lng2 = np.sqrt(ode_rhs2(v_grid, t_grid)**2 + 1) #Lenght of the vector
dt2 = 1/lng2 #get the horizontal component
dv2 = ode_rhs2(v_grid, t_grid) / lng #get the vertical component
```

```python
iso_line2 = {}
for i, iso in enumerate(ISO):
     iso_line2[i]= ode_rhs2(iso, t)
```

```python
fig, ax = plt.subplots(figsize=(10, 10));
ax.set_ylabel('dV/dt')
ax.set_xlabel('Time (s)')
ax.set_title('')
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

plt.yticks(np.arange(-5, 6, 1));
plt.xticks(np.arange(-5, 6, 1));
plt.quiver(t, v, dt2, dv2);

iso_plt = []
iso_plt0, = plt.plot(t, iso_line2[0], linewidth=2, label='-2')
iso_plt1, = plt.plot(t, iso_line2[1], linewidth=2, label='-1')
iso_plt2, = plt.plot(t, iso_line2[2], linewidth=2, label=' 0')
iso_plt3, = plt.plot(t, iso_line2[3], linewidth=2, label=' 1')
iso_plt4, = plt.plot(t, iso_line2[4], linewidth=2, label=' 2')
plt.legend(loc=1, handles=[iso_plt0, iso_plt1, iso_plt2, iso_plt3, iso_plt4])

plt.savefig('report/1_0.png', dpi=200)
```

## Differential equations of a simple cell model

```python
# tau = R * C
# tau * du/dt = urest - u + R * I 
```

### R = 1, C =1, Imax = 0

```python
R, C = 1, 1
```

```python
t = np.linspace(-5, 5, 30)
v = np.linspace(-5, 5, 30)
```

```python
I_cal = lambda t, Imax: Imax * np.sin(t)
```

```python
ode_rhs_cell = lambda V, I: (-V + R * I)/ R * C
```

```python
I0 = I_cal(t, 0)
I0_grid , v_grid0 = np.meshgrid(I0, v)
```

```python
lng0 = np.sqrt(ode_rhs_cell(v_grid0, I0_grid)**2 + 1) #Lenght of the vector
dI0 = 1/lng0 #get the horizontal component
dv0 = ode_rhs_cell(v_grid0, I0_grid) / lng0 #get the vertical component
```

```python
iso_line3 = {}
for i, iso in enumerate(ISO):
     iso_line3[i]= ode_rhs_cell(iso, I0)
```

```python
fig, ax = plt.subplots(figsize=(10, 10));
ax.set_ylabel('dV/dt')
ax.set_xlabel('Time (s)')
ax.set_title('')
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

plt.yticks(np.arange(-5, 6, 1));
plt.xticks(np.arange(-5, 6, 1));
plt.quiver(t, v, dI0, dv0);



# for iso, n in enumerate(iso_line3):
#     plt.plot(t, iso_line3[n], linewidth=2)
    
iso_plt3 = []
iso_plt30, = plt.plot(t, iso_line3[0], linewidth=2, label='-2')
iso_plt31, = plt.plot(t, iso_line3[1], linewidth=2, label='-1')
iso_plt32, = plt.plot(t, iso_line3[2], linewidth=2, label=' 0')
iso_plt33, = plt.plot(t, iso_line3[3], linewidth=2, label=' 1')
iso_plt34, = plt.plot(t, iso_line3[4], linewidth=2, label=' 2')
plt.legend(loc=1, handles=[iso_plt30, iso_plt31, iso_plt32, iso_plt33, iso_plt34])

plt.savefig('report/2_0.png', dpi=200)
```
