## Forward Euler

```python
v0 = -4
t0 = -4.5

def f(V,t):
    return 1 - V - t

def euler(v0, t0, f, dt, T):
    t = np.linspace(t0, T, int(T/dt)+1)
    V = np.zeros(len(t))
    V[0] = v0
    for i in np.arange(1, len(t)):
        V[i] = V[i-1] + f(V[i-1], t[i-1])* dt
        
    return V, t
```

```python
V0, T0 = euler(v0, t0, f, 1, 5)
V1, T1 = euler(v0, t0, f, 0.5, 5)
V2, T2 = euler(v0, t0, f, 0.1, 5)
V3, T3 = euler(v0, t0, f, 0.012, 5)

fig, ax = plt.subplots(figsize=(20, 10));
ax.set_ylabel('V (V)')
ax.set_xlabel('t (s)')
ax.grid()
ax.set_xlim(-5, 5.5)
ax.set_ylim(-5, 6)

plt.yticks(np.arange(-5, 6, 1));
plt.xticks(np.arange(-5, 6, 1));

l0, = plt.plot(T0, V0, label='dt = 1s')
l1, = plt.plot(T1, V1, label='dt = 0.5s')
l2, = plt.plot(T2, V2, label='dt = 0.1s')
l3, = plt.plot(T3, V3, label='dt = 0.012s')

plt.legend(loc=1, handles=[l0, l1, l2, l3])
```

## Heun Method
modified Euler, has an additional intermediete state

```python
def heun(v0, t0, f, dt, T):
    t = np.linspace(t0, T, int(T/dt)+1)
    V = np.zeros(len(t))
    V[0] = v0
    for i in np.arange(1, len(t)):
        A = f(V[i-1], t[i-1])
        Vp = V[i-1] + A * dt
        B = f(Vp, t[i])
        V[i] = V[i-1] + (A+B)/2 * dt
        
    return V, t
```

```python
hV0, T0 = heun(v0, t0, f, 1, 5)
hV1, T1 = heun(v0, t0, f, 0.5, 5)
hV2, T2 = heun(v0, t0, f, 0.1, 5)
hV3, T3 = heun(v0, t0, f, 0.012, 5)

fig, ax = plt.subplots(figsize=(20, 10));
ax.set_ylabel('V (V)')
ax.set_xlabel('t (s)')
ax.grid()
ax.set_xlim(-5, 5.5)
ax.set_ylim(-5, 6)

plt.yticks(np.arange(-5, 6, 1));
plt.xticks(np.arange(-5, 6, 1));

l0, = plt.plot(T0, hV0, label='dt = 1s')
l1, = plt.plot(T1, hV1, label='dt = 0.5s')
l2, = plt.plot(T2, hV2, label='dt = 0.1s')
l3, = plt.plot(T3, hV3, label='dt = 0.012s')

plt.legend(loc=1, handles=[l0, l1, l2, l3])
```

## Exponential Euler

```python
def exp_euler(v0, t0, f, dt, T):
#     f = A(t) V(t) + (B(V,t))
#     f = 1 - V - t
    t = np.linspace(t0, T, int(T/dt)+1)
    V = np.zeros(len(t))
    V[0] = v0
    for i in np.arange(1, len(t)):
        A = -1
        B = 1 - t[i]
        V[i] = V[i-1] * np.exp(A*dt) + B/A * (np.exp(A*dt) - 1) 
        
    return V, t
```

```python
eV0, T0 = exp_euler(v0, t0, f, 1, 5)
eV1, T1 = exp_euler(v0, t0, f, 0.5, 5)
eV2, T2 = exp_euler(v0, t0, f, 0.1, 5)
eV3, T3 = exp_euler(v0, t0, f, 0.012, 5)

fig, ax = plt.subplots(figsize=(20, 10));
ax.set_ylabel('V (V)')
ax.set_xlabel('t (s)')
ax.grid()
ax.set_xlim(-5, 5.5)
ax.set_ylim(-5, 6)

plt.yticks(np.arange(-5, 6, 1));
plt.xticks(np.arange(-5, 6, 1));

l0, = plt.plot(T0, eV0, label='dt = 1s')
l1, = plt.plot(T1, eV1, label='dt = 0.5s')
l2, = plt.plot(T2, eV2, label='dt = 0.1s')
l3, = plt.plot(T3, eV3, label='dt = 0.012s')

plt.legend(loc=1, handles=[l0, l1, l2, l3])
```

## The Leaky Integrate and Fire Neuron

### Model for constant currents

```python
def LIF(T, dt, I):
    Cm = 1e-6
    g_leak = 100e-6
    V_rest  = -60e-3
    V_thr   = -20e-3
    V_spike = 20e-3

    t = np.linspace(0, T, int(T/dt))
    V = np.zeros(len(t))
    V[0] = V_rest

    for n in np.arange(0, len(t)-1):
        if   (V[n] < V_thr):
            V[n+1] = V[n] + dt/Cm * (-g_leak * (V[n] - V_rest) + I)
        elif (V[n] == V_spike):
            V[n+1] = V_rest
        elif (V[n] >= V_thr):
            V[n+1] = V_spike

            
    return V, t
```

## Input Current of 10uA

```python
I = 10e-6
T= 50e-3
dt = 25e-6

Vl, tl = LIF(T, dt, I)

fig, ax = plt.subplots(figsize=(20, 10));
ax.set_ylabel('V (V)')
ax.set_xlabel('t (s)')
ax.grid()
# ax.set_xlim(0, 50)
# ax.set_ylim(-60, 20)
# plt.yticks(np.arange(-5, 6, 1));
# plt.xticks(np.arange(-5, 6, 1));

plt.plot(tl, Vl);
```

## Rectified Sine-Input 

```python
def sin_rect(f, dt, T, A):
    w = 2. * np.pi * f
    t = np.linspace(0, T, int(T/dt))
    s = np.abs(A * np.sin(w*t))

    return s, t
```

```python
T= 50
dt = 0.05
f = 50
s, t = sin_rect(f, dt, T, 10)
fig, ax = plt.subplots(figsize=(10, 10));
ax.set_ylabel('I (uA)')
ax.set_xlabel('t (ms)')
ax.grid()
plt.plot(t, s);

s2, t2 = sin_rect(f, dt, T, 30)

fig, ax = plt.subplots(figsize=(10, 10));
ax.set_ylabel('I (uA)')
ax.set_xlabel('t (ms)')
ax.grid()
plt.plot(t2, s2);
```

## Applying rectified sine to the LIF 

### Model adaptation

```python
def LIFsin(T, dt, Isin):
    Cm = 1e-6
    g_leak = 100e-6
    V_rest  = -60e-3
    V_thr   = -20e-3
    V_spike = 20e-3

    t = np.linspace(0, T, int(T/dt))
    V = np.zeros(len(Isin))
    V[0] = V_rest

    for n in np.arange(0, len(Isin)-1):
        if   (V[n] < V_thr):
            V[n+1] = V[n] + dt/Cm * (-g_leak * (V[n] - V_rest) + Isin[n]*10e-7)
        elif (V[n] == V_spike):
            V[n+1] = V_rest
        elif (V[n] >= V_thr):
            V[n+1] = V_spike

    return V, t
```

```python
T= 50e-3
dt = 25e-6

Isin10, t2 = sin_rect(f, dt, T, 10)
V_sin, t_sin = LIFsin(T, dt, Isin10)

print(t2)
print(t_sin)

fig, ax = plt.subplots(figsize=(20, 10));
ax.set_ylabel('V (V)')
ax.set_xlabel('t (s)')
ax.grid()
plt.plot(t_sin, V_sin);
```
